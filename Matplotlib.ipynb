```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(4,15,100)
y = np.sin(x)
```

```python
plt.figure(figsize=(4,2))
plt.title("Line Plot")
plt.xlabel("x values")
plt.ylabel("Y values")
plt.grid(True)
plt.plot(x, y, "k--",label = "Line")
plt.legend()
plt.show()
```

```python
y= np.array([4,6,3,5,7])
x = np.array(["A","B","C","D","E"])
```

```python
plt.figure(figsize=(4,2))
plt.title("Bar Chart")
plt.xlabel("x values")
plt.ylabel("Y values")
#plt.grid(True)
plt.bar(x, y,label = "Bar")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Bar Chart")
plt.xlabel("x values")
plt.ylabel("Y values")
plt.bar(x, y,color = "Black")
plt.legend(["Bar"])
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
size = 1000 * np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
plt.title("Scatter Plot")
plt.xlabel("x values")
plt.ylabel("Y values")
plt.scatter(x, y,c=colors, s=size, cmap="viridis",alpha = 0.5)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Scatter Plot")
plt.xlabel("x values")
plt.ylabel("Y values")
plt.scatter(x, y,c=colors, s=size, cmap="plasma",alpha = 0.5)
plt.colorbar(label = "Intensity")
plt.show()
```

```python
data = np.random.normal(size=1000)
plt.figure(figsize=(5,3))
plt.title("Histogram")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.hist(data, histtype="bar",bins=30,color ="violet",edgecolor="Black",alpha=0.7)
plt.show()
```

```python
size = np.array([34,56,21,69])
labels = np.array(["Ram","Sham","Sita","Gita"])
```

```python
plt.figure(figsize=(5,4))

plt.pie(size, labels=labels, explode=[0.0,0.0,0.1,0.0], autopct ="%1.1f%%", 
        colors = (["r","g","b","y"]))
plt.title("Pie Chart")
plt.axis("equal")
plt.show()
```

```python
plt.figure(figsize=(5,4))
cols = ["cyan", "teal", "turquoise", "skyblue"]
plt.pie(size, labels=labels, explode=[0.2,0.0,0.0,0.0], autopct ="%1.1f%%", 
        colors = cols)
plt.title("Pie Chart")
plt.axis("equal")
plt.show()
```

```python
angle = np.linspace(0,2*np.pi,100)
radii = np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
ax = plt.subplot(111,polar=True)
ax.plot(angle, radii,color="red",linewidth=2)
ax.fill(angle, radii,color="red",alpha=0.3)
plt.title("polar plot")
plt.show()
```
