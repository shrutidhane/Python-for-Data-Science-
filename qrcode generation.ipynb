```python
#import librabries
import qrcode
import matplotlib.pyplot as plt
```

```python
#Add data which you want to be QR code representation
data = "http://www.unipune.ac.in/"

#Output file name 
filename ="My_QR_Code.png"

#Generate qr code
image = qrcode.make(data)

#save image into a file 
image.save(filename)

#show qr code 
plt.imshow(image,cmap='gray')
```
