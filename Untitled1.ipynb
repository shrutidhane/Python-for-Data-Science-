```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.datasets import make_regression
```

```python
#Generating data using make_regression

data = make_regression(n_samples=1000,n_features=1,noise=20)
```

```python
X = data[0]
Y = data[1]
```

```python
df = pd.DataFrame(X,columns=["Feature1"])
df["Target"] =Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=57)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```
